```python
class vehicle:
  def __init__(self,id,brand,rent):
    self.id=id
    self.brand=brand
    self.rent=rent

  def disp_details(self):
    print("id:",self.id)
    print("brand:",self.brand)
    print("rent:",self.rent)

  def calcrent(self,day):
    return self.rent*day


mehran=vehicle(100,"SUZUKI",1000)
mehran.disp_details()
alto=vehicle(101,"TOYOTA",1500)
alto.disp_details()
print("rent:",mehran.calcrent(5))
print("rent:",alto.calcrent(5))
```

```python
class student:
  def __init__(self,name,id,marks):
    self.name=name
    self.id=id
    self.__marks=marks

  def setmarks(self,marks):
    self.__marks=marks

  def getmarks(self):
    return self.__marks

  def calcgrade(self):
    if self.__marks >=80:
      return "A"
    elif self.__marks >=70 and self.__marks < 80:
      return "B"
    elif self.__marks >=60 and self.__marks < 70:
      return "C"
    elif self.__marks >=50 and self.__marks < 60:
      return "D"
    else :
      return "fail"


usman = student("Usman",100,100)
print(usman.getmarks())
print(usman.calcgrade())

huzaifa=student("huzaifa",101,80)
huzaifa.setmarks(40)
print(huzaifa.getmarks())
print(huzaifa.calcgrade())
```
