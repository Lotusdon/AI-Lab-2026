```python
import random

class environment:
  def __init__(self):
    self.present="No"
    self.light="OFF"

  def update(self):
    self.present=random.choice(["Yes","No"])

  def getp(self):
    return self.present,self.light

  def setlight(self,state):
    self.light=state


class agent:
  def __init__(self):
    self.prev_present="No"
    self.prev_light="OFF"

  def act(self,percept,env):
    present,light=percept

    if present=="Yes" and light=="OFF":
      env.setlight("ON")
      action="Turn lights ON"
    elif present=="No" and light=="ON":
      env.setlight("OFF")
      action="Turn lights OFF"
    else:
      action="No action"

    self.prev_present=present
    self.prev_light=env.light

    return action


def run(agent,env):
  for i in range(8):
    env.update()
    percept=env.getp()
    action=agent.act(percept,env)

    print("Step",i+1)
    print("Students :",percept[0],", Light :",percept[1])
    print("Action :",action)
    print("Model : Prev Students:",agent.prev_present,"Prev Light:",agent.prev_light)
    print()


agent=agent()
env=environment()

run(agent,env)
```

```python
class environment:
  def __init__(self):
    self.sub=["AI","Math","Physics"]
    self.done=[False,False,False]

  def getp(self):
    return self.sub,self.done

  def complete(self,i):
    self.done[i]=True


class agent:
  def __init__(self):
    pass

  def act(self,percept,env):
    sub,done=percept

    for i in range(len(sub)):
      if done[i]==False:
        print("Studying",sub[i])
        env.complete(i)
        return

    print("Goal Achieved! All subjects completed")


def run(agent,env):
  while True:
    percept=env.getp()
    sub,done=percept

    if all(done):
      print("Goal Achieved! All subjects completed")
      break

    agent.act(percept,env)


agent=agent()
env=environment()

run(agent,env)
```

```python
class environment:
  def __init__(self):
    self.name=["A","B"]
    self.dist=[3,5]
    self.rate=[7,9]

  def getp(self):
    return self.name,self.dist,self.rate


class agent:
  def __init__(self):
    pass

  def act(self,percept):
    name,dist,rate=percept

    util=[]
    for i in range(len(name)):
      u=rate[i]-dist[i]
      util.append(u)
      print("Restaurant",name[i],"Utility =",u)

    best=0
    for i in range(1,len(util)):
      if util[i]>util[best]:
        best=i

    print("Selected Restaurant:",name[best])


def run(agent,env):
  percept=env.getp()
  agent.act(percept)


agent=agent()
env=environment()

run(agent,env)
```

```python
import random

class agent:
  def __init__(self):
    self.q={"Play":0,"Rest":0}
    self.lr=0.1
    self.r={"Play":5,"Rest":1}

  def update(self,a):
    self.q[a]=self.q[a]+self.lr*(self.r[a]-self.q[a])

  def run(self,steps=10):
    for i in range(steps):
      a=random.choice(list(self.r.keys()))
      rew=self.r[a]
      self.update(a)

      print("Step",i+1)
      print("Action:",a,",Reward:+"+str(rew))

    print("Q-table Updated")
    print(self.q)


agent=agent()
agent.run()
```

```python
class environment:
  def __init__(self):
    self.grid={
      "a":"safe","b":"safe","c":"fire",
      "d":"safe","e":"fire","f":"safe",
      "g":"safe","h":"safe","j":"fire"
    }

  def get(self,room):
    return self.grid[room]

  def put(self,room,state):
    self.grid[room]=state

  def show(self):
    rooms=["a","b","c","d","e","f","g","h","j"]
    for i in range(0,9,3):
      row=""
      for r in rooms[i:i+3]:
        if self.grid[r]=="fire":
          row=row+" F "
        else:
          row=row+" . "
      print(row)
    print()


class agent:
  def __init__(self):
    self.path=["a","b","c","d","e","f","g","h","j"]

  def run(self,env):
    for r in self.path:
      print("Robot in room",r)

      if env.get(r)=="fire":
        print("Fire detected → Extinguishing")
        env.put(r,"safe")
      else:
        print("No fire")

      env.show()

    print("Final Status:")
    env.show()


agent=agent()
env=environment()

agent.run(env)
```
